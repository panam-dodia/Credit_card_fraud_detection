```python
df.drop_duplicates()

df
```

```python
# List of columns to drop from the DataFrame
columns_to_drop = ["Transaction Date and Time", "Cardholder Name", "Card Number", "Merchant Name", "Transaction Location", "Card Expiration Date", "CVV Code", "Transaction ID", "IP Address", "User Account Information", "Transaction Notes",]
df.drop(columns=columns_to_drop, inplace=True)

df
```

```python
# Mapping old column names to new column names
new_column_names = {
    'Transaction Amount': 'transaction_amount',
    'Merchant Category Code' : 'merchant_category_code',
    'Transaction Currency' : 'transaction_currency',
    'Card Type' : 'card_type',
    'Transaction Response Code' : 'transaction_response_code',
    'Fraud Flag or Label' : 'fraud_flag',
    'Previous Transactions' : 'previous_transactions',
    'Transaction Source' : 'transaction_source',
    'Device Information' : 'device_information'
}
df.rename(columns=new_column_names, inplace=True)
```

```python
# Filling missing values in the DataFrame with 0
df.fillna(0, inplace=True)
df['previous_transactions'] = df['previous_transactions'].replace('3 or more', 3)
df['previous_transactions'] = df['previous_transactions'].astype(int)

df
```

```python
# Initialize an instance of the OrdinalEncoder class
encode = OrdinalEncoder()
df[["previous_transactions"]] = encode.fit_transform(df[["previous_transactions"]])

df
```

```python
encoder = OneHotEncoder()

# Fit and transform the data
onehot_encoded = encoder.fit_transform(df[['device_information', 'transaction_source', 'card_type']])
onehot_encoded_df = pd.DataFrame(onehot_encoded.toarray(), columns=encoder.get_feature_names_out(['device_information', 'transaction_source', 'card_type']))
df_encoded = pd.concat([df, onehot_encoded_df], axis=1)

# Drop the original categorical columns
df_encoded.drop(['device_information', 'transaction_source', 'card_type'], axis=1, inplace=True)

df = df_encoded
df
```

```python
exchange_rates = {'USD': 1.0, 'INR': 0.014, 'EUR': 1.18}

# Convert amounts to USD
df['transaction_amount'] = df.apply(lambda row: row['transaction_amount'] * exchange_rates[row['transaction_currency']], axis=1)

df
```

```python
df.drop("transaction_currency", axis=1, inplace = True)

df
```

```python
df['merchant_category_code'] = df['merchant_category_code'].astype(int)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
import numpy as np

# Define the ANN architecture
class FraudDetectionNN(nn.Module):
    def __init__(self, input_size):
        super(FraudDetectionNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.sigmoid(self.fc3(x))
        return x

# Define function to train the model
def train_model(model, criterion, features, labels, num_epochs=10, batch_size=64, learning_rate=0.01):
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

    for epoch in range(num_epochs):
        running_loss = 0.0
        for i in range(0, len(features), batch_size):
            inputs = torch.FloatTensor(features[i:i+batch_size])
            targets = torch.FloatTensor(labels[i:i+batch_size]).unsqueeze(1)

            optimizer.zero_grad()

            outputs = model(inputs)
            loss = criterion(outputs, targets)

            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)

        epoch_loss = running_loss / len(features)
        print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")
```

```python
if __name__ == "__main__":
    labels = df['fraud_flag'].values
    features = df.drop(columns=['fraud_flag']).values

    # Split data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

    input_size = X_train.shape[1]  
    model = FraudDetectionNN(input_size)
    criterion = nn.BCELoss()  
    optimizer = optim.Adam(model.parameters(), lr=0.001) 

    train_model(model, criterion, X_train, y_train, num_epochs=10, batch_size=64)

    # Evaluate the model on test set
    model.eval()
    with torch.no_grad():
        inputs = torch.FloatTensor(X_test)
        targets = torch.FloatTensor(y_test).unsqueeze(1)
        outputs = model(inputs)
        test_loss = criterion(outputs, targets)
        print(f"Test Loss: {test_loss.item():.4f}")
```
